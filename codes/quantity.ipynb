```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv("/content/drive/MyDrive/quantity/quantity_by_month1.csv/part-00000-f8881a68-6120-46a8-9d00-ed750c6f44b5-c000.csv")

data.dropna(inplace=True)

X = data[['Month', 'Year', 'Order Region']]
y = data['sum(Order Item Quantity)']

predictions_dict = {}
for product_name in data['Product Name'].unique():
    product_data = data[data['Product Name'] == product_name]

    X_train = product_data[['Month', 'Year', 'Order Region']]
    y_train = product_data['sum(Order Item Quantity)']
    X_train = pd.get_dummies(X_train, columns=['Order Region'], drop_first=True)

    model = LinearRegression()
    model.fit(X_train, y_train)

    for month in range(1, 13):
        for region in data['Order Region'].unique():
            prediction_features = pd.DataFrame({
                'Month': [month],
                'Year': [2019],
                'Order Region': [region]
            })
            merged_data = pd.concat([X_train, prediction_features])
            merged_data_encoded = pd.get_dummies(merged_data, columns=['Order Region'], drop_first=True)
            X_train_encoded = merged_data_encoded.iloc[:len(X_train)]
            prediction_features_encoded = merged_data_encoded.iloc[len(X_train):]

            prediction_features_encoded.fillna(0, inplace=True)
            predicted_quantity = model.predict(prediction_features_encoded)

            predictions_dict[(product_name, month, region)] = predicted_quantity[0]


for key, value in predictions_dict.items():
    print(f"Predicted quantity for product '{key[0]}' in {key[2]} for month {key[1]} in 2019:", value)
predictions_df = pd.DataFrame(predictions_dict.values(), index=predictions_dict.keys(), columns=['Predicted_Quantity'])
predictions_df.reset_index(inplace=True)
predictions_df.columns = ['Product_Name', 'Month', 'Order_Region', 'Predicted_Quantity']
predictions_df.to_csv("/content/drive/MyDrive/quantity/predictions.csv", index=False)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
y_pred_train = model.predict(X_train_encoded)

rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
print("RMSE on training set:", rmse_train)
from sklearn.metrics import mean_absolute_error
mae_train = mean_absolute_error(y_train, y_pred_train)
print("MAE on training set:", mae_train)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

data = pd.read_csv("/content/drive/MyDrive/quantity/quantity_by_month1.csv/part-00000-f8881a68-6120-46a8-9d00-ed750c6f44b5-c000.csv")

data.dropna(inplace=True)

X = data[['Month', 'Year', 'Order Region']]
y = data['sum(Order Item Quantity)']

predictions_dict = {}

for product_name in data['Product Name'].unique():
    product_data = data[data['Product Name'] == product_name]

    X_train = product_data[['Month', 'Year', 'Order Region']]
    y_train = product_data['sum(Order Item Quantity)']

    X_train = pd.get_dummies(X_train, columns=['Order Region'], drop_first=True)

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    for month in range(1, 13):
        for region in data['Order Region'].unique():
            prediction_features = pd.DataFrame({
                'Month': [month],
                'Year': [2019],
                'Order Region': [region]
            })
            merged_data = pd.concat([X_train, prediction_features])
            merged_data_encoded = pd.get_dummies(merged_data, columns=['Order Region'], drop_first=True)
            X_train_encoded = merged_data_encoded.iloc[:len(X_train)]
            prediction_features_encoded = merged_data_encoded.iloc[len(X_train):]
            prediction_features_encoded.fillna(0, inplace=True)
            predicted_quantity = model.predict(prediction_features_encoded)
            predictions_dict[(product_name, month, region)] = predicted_quantity[0]

for key, value in predictions_dict.items():
    print(f"Predicted quantity for product '{key[0]}' in {key[2]} for month {key[1]} in 2019:", value)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
y_pred_train_rf = model.predict(X_train_encoded)
rmse_train_rf = np.sqrt(mean_squared_error(y_train, y_pred_train_rf))
print("RMSE on training set (Random Forest):", rmse_train_rf)
from sklearn.metrics import mean_absolute_error
mae_train_rf = mean_absolute_error(y_train, y_pred_train_rf)
print("MAE on training set (Random Forest):", mae_train_rf)
```

```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

data = pd.read_csv("/content/drive/MyDrive/quantity/quantity_by_month1.csv/part-00000-f8881a68-6120-46a8-9d00-ed750c6f44b5-c000.csv")

data.dropna(inplace=True)

X = data[['Month', 'Year', 'Order Region']]
y = data['sum(Order Item Quantity)']

predictions_dict = {}
for product_name in data['Product Name'].unique():
    product_data = data[data['Product Name'] == product_name]
    X_train = product_data[['Month', 'Year', 'Order Region']]
    y_train = product_data['sum(Order Item Quantity)']
    X_train = pd.get_dummies(X_train, columns=['Order Region'], drop_first=True)
    model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
    model.fit(X_train, y_train)

    for month in range(1, 13):
        for region in data['Order Region'].unique():
            prediction_features = pd.DataFrame({
                'Month': [month],
                'Year': [2019],
                'Order Region': [region]
            })
            merged_data = pd.concat([X_train, prediction_features])
            merged_data_encoded = pd.get_dummies(merged_data, columns=['Order Region'], drop_first=True)
            X_train_encoded = merged_data_encoded.iloc[:len(X_train)]
            prediction_features_encoded = merged_data_encoded.iloc[len(X_train):]
            prediction_features_encoded.fillna(0, inplace=True)
            predicted_quantity = model.predict(prediction_features_encoded)
            predictions_dict[(product_name, month, region)] = predicted_quantity[0]

for key, value in predictions_dict.items():
    print(f"Predicted quantity for product '{key[0]}' in {key[2]} for month {key[1]} in 2019:", value)

y_pred_train_gbm = model.predict(X_train_encoded)

rmse_train_gbm = np.sqrt(mean_squared_error(y_train, y_pred_train_gbm))
print("RMSE on training set (Gradient Boosting):", rmse_train_gbm)

mae_train_gbm = mean_absolute_error(y_train, y_pred_train_gbm)
print("MAE on training set (Gradient Boosting):", mae_train_gbm)
predictions_df = pd.DataFrame(predictions_dict.values(), index=predictions_dict.keys(), columns=['Predicted_Quantity'])

predictions_df.reset_index(inplace=True)
predictions_df.columns = ['Product_Name', 'Month', 'Order_Region', 'Predicted_Quantity']
predictions_df.to_csv("/content/drive/MyDrive/quantity/predictions_rf.csv", index=False)
```
